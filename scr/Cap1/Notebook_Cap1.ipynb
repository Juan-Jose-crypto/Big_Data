```python
path_1 = '../../data/Cap1/subvenciones.csv'
path_2 = '../../data/Cap1/subvenciones_esc.csv'
```

```python
with open(path_2, encoding = 'latin1') as file_write:
    reader = csv.reader(file_write)

    for row in reader:
        print(row)
```

```python
with open(path_1, encoding = 'latin1') as file_read, open(path_2, 'w', encoding = 'latin1') as file_write:
    reader = csv.DictReader(file_read)
    headers = reader.fieldnames + ['Justificación requerida', 'Justificación recibida']
    writer = csv.DictWriter(file_write, fieldnames = headers)

    writer.writeheader()
    for row in reader:
        if float(row['Importe']) > 300:
            row['Justificación requerida'] = 'Sí'
        else:
            row['Justificación requerida'] = 'No'

        row['Justificación recibida'] = 'No'

        writer.writerow(row)
```
